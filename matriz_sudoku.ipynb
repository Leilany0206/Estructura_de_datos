```python
sudoku = [
  [5, 3, 0, 0, 7, 0, 0, 0, 0],
  [6, 0, 0, 1, 9, 5, 0, 0, 0],
  [0, 9, 8, 0, 0, 0, 0, 6, 0],
  [8, 0, 0, 0, 6, 0, 0, 0, 3],
  [4, 0, 0, 8, 0, 3, 0, 0, 1],
  [7, 0, 0, 0, 2, 0, 0, 0, 6],
  [0, 6, 0, 0, 0, 0, 2, 8, 0],
  [0, 0, 0, 4, 1, 9, 0, 0, 5],
  [0, 0, 0, 0, 8, 0, 0, 7, 9]
  ]

originalSudoku = [
  [5, 3, 0, 0, 7, 0, 0, 0, 0],
  [6, 0, 0, 1, 9, 5, 0, 0, 0],
  [0, 9, 8, 0, 0, 0, 0, 6, 0],
  [8, 0, 0, 0, 6, 0, 0, 0, 3],
  [4, 0, 0, 8, 0, 3, 0, 0, 1],
  [7, 0, 0, 0, 2, 0, 0, 0, 6],
  [0, 6, 0, 0, 0, 0, 2, 8, 0],
  [0, 0, 0, 4, 1, 9, 0, 0, 5],
  [0, 0, 0, 0, 8, 0, 0, 7, 9]
  ]

while True: #inicializa el while

  # SIMULACIÓN DE PPRINT
  def pprint():
    for i in sudoku:
      print(i)

  # Declara valores
  row = 0
  column = 0
  value = 0

  # INPUTS
  # Filas
  def askRow():
    while True:
      try:
        row = int(input('Ingresa la fila (1-9) o 0 para salir: '))

        if not (0 <= row <= 9):
          print('Por favor ingresa un valor entre 1 y 9 o 0 para salir')
          pass
        else:
          return(row)
          break

      except ValueError:
        print('Por favor ingresa un dato válido númerico')

  pprint()
  row = askRow()
  if row == 0:
    break

  # Columnas
  def askColumn():
    while True:
      try:
        column = int(input('Ingresa la columna (1-9) o 0 para salir: '))

        if not (0 <= column <= 9):
          print('Por favor ingresa un valor entre 1 y 9 o 0 para salir')
          pass
        else:
          return(column)
          break

      except ValueError:
        print('Por favor ingresa un dato válido númerico')

  column = askColumn()
  if column == 0:
    break

  # Valores
  def askValue():
    while True:
      try:
        value = int(input('Ingrese un valor (1-9): '))

        if not (1 <= value <= 9):
          print('Por favor ingresa un valor entre 1 y 9')
          pass
        else:
          return(value)
          break

      except ValueError:
        print('Por favor ingresa un dato válido númerico')

  value = askValue()

  # FUNCIÓN QUE CÁLCULA CADA CUADRANTE
  def checkPart():

    auxRow = [] # Incializa un auxiliar que verifica la línea

    if row % 3 == 0:  # Si la línea es multiplo de 3 (3, 6 o 9)
      for i in range(row-2, (row+1), 1): # Ciclo de i++ que comienza en 2 menos hasta 1 más (2 - linea + 1)
        auxRow.append(i) # Guarda el rango del "cuadrante"
    elif row == 1 or row == 4 or row == 7: # Si la línea es 1, 4 o 3
      for i in range(row, (row+3), 1):  # Ciclo de i++ que termina en 3 más (linea + 3)
        auxRow.append(i)
    elif row == 2 or row == 5 or row == 8: # Si la línea es 2, 5 o 8
      for i in range(row-1, (row+2), 1): # Ciclo de i++ que comienza en 1 menos hasta 2 más (1 - linea + 2)
        auxRow.append(i)

    auxColumn = [] # Incializa un auxiliar que verifica la columna

    # Misma lógica que en row
    if column % 3 == 0:
        for i in range(column-2, (column+1), 1):
            auxColumn.append(i)
    elif column == 1 or column == 4 or column == 7:
        for i in range(column, (column+3), 1):
            auxColumn.append(i)
    elif column == 2 or column == 5 or column == 8:
        for i in range(column-1, (column+2), 1):
            auxColumn.append(i)

    auxVerify = [] # Inicializa una lista auxiliar para guardar el cuadrante

    for i in auxRow: # Bucle que recorre el rango del cuadrante por línea obtenido arriba
      for x in auxColumn: # Bucle que recorre el rango del cuadrante por columna obtenido arriba
        auxVerify.append(sudoku[i-1][x-1]) #Une por cada iteración los valores del cuadrante

    return auxVerify

  # FUNCIÓN PARA CALCULAR CADA COLUMNA
  def checkColumn():

    auxCheckColumn = [] # Inicializa una lista auxiliar para guardar la columna

    for i in range(0, 9, 1): # Iteración del 0 al 9
      auxCheckColumn.append(sudoku[i][column-1]) # Agrega a la lista auxiliar cada valor en la posición de columna de cada línea

    return auxCheckColumn

  # Lógica del sudoku
  def verify():

    while True:
      if value in sudoku[row-1]: # Busca si el valor se encuntra en la línea
        print(f'{value} ya está en la fila {row}')
        break
      elif value in checkColumn(): # Busca si el valor se encuntra en la columna con ayuda de la lista generada en checkColumn()
        print(f'{value} ya está en la columna {column}')
        break
      elif value in checkPart(): # Busca si el valor se encuentra en la lista que contiene los valores del cuadrante generada en checkPart()
        print(f'{value} ya está en ese cuadrante')
        break
      elif originalSudoku[row-1][column-1] == 0: # Si el sudoku original es 0 entonces se puede reasignar valores en 0
        sudoku[row-1][column-1] = value
        break
      elif originalSudoku[row-1][column-1] != 0: # Busca si el sudoku original admite que se cambie ese valor
        print('No puedes modificar un valor original')
        break

  verify()

  # REVISIÓN DE QUE SE HAYA COMPLTEADO EL SUDOKU
  for x in range(0, 9, 1):
    if 0 not in sudoku[x]:
      print('¡FELICIDADES! ¡RESOLVISTE EL SUDOKU!')
      break
```
